### 理解を深める
- 偏差    
「個々のデータ」から「平均」を引いたもの
- 平方和  
【目的】平均を基準としてデータの散らばり具合を数値化したもの
$$
（個々のデータ　-　平均）^2の和
$$
- 分散   
【目的】どれだけみんなの値がバラけているか調べる
$$
\frac{平方和}{データの個数}
$$
- 標準偏差  
【目的】単位を元通りにした指標
$$
\sqrt{分散}
$$
- 共分散  
【目的】二つのデータについて、片方が増えているとき、もう片方は増えている（あるいは減っている）のか？を調べる
「国語の点数X」が高いほど「数学の点数Y」が高い傾向にあるのか？あるいは「国語の点数X」と「数学の点数Y」は関係ないのか？などが分析できる。
5人の「国語の点数X」と「数学の点数Y」の共分散の場合
$$
\frac{国語の点数Xの偏差\times数学の点数Yの偏差}{5人}
$$
共分散は，「X の偏差 × Y の偏差」の平均なので<br>
共分散が大きい（正）→ X が大きいとき Y も大きい傾向がある<br>
共分散が 0 に近い→ X と Y にあまり関係はない<br>
共分散が小さい（負）→ X が大きいとき Y は小さい傾向がある<br>
と言える

共分散は英語で「Covariance」と言うので，X と Y の共分散のことを Cov(X,Y)やσXY と書くことがある
$$
Cov(X, Y) = E[(X - μ_X)(Y - μ_Y)]  \\
Eは期待値（平均）、μ_X、μ_Yは平均
$$
- 分散共分散行列  
 【目的】分散（散らばり具合を表す指標）の概念を多次元確率変数に拡張して行列としたもの
 $$
 \sigma^2 は分散を表す\\
 \sum= \left(
\begin{matrix} 
\sigma^2_X & \sigma_{XY} \\ 
\sigma_{XY} & \sigma^2_Y
\end{matrix} 
\right)=
 \sum= \left(
\begin{matrix} 
Xの分散 & 共分散 \\ 
共分散 & Yの分散
\end{matrix} 
\right)
$$

### 正規分布
- 正規分布は確率密度関数の中のひとつ（確率密度関数にはt分布やF分布などもある）
- 正規分布は平均を境に左右対称である
- 正規分布は平均と標準偏差の影響をうける
- 正規分布の確率密度関数は下記の形（ガウス関数と呼ばれる）
$$
f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}exp\bigl(-\frac{(x - \mu)^2}{2\sigma^2}\bigr)
$$
#### 標準正規分布
- 平均が0で標準偏差が1の正規分布のことを標準正規分布という
$$
平均が53で標準偏差が10の正規分布\\
f(x)=\frac{1}{\sqrt{2\pi\times10}}exp\bigl(-\frac{(x - 53)^2}{2\times10}\bigr)\\
平均が0で標準偏差が1の標準正規分布\\
f(x)=\frac{1}{\sqrt{2\pi\times1}}exp\bigl(-\frac{(x - 0)^2}{2\times1}\bigr)\\
$$
#### n次元正規分布
- n次元正規分布（多変量正規分布）の同時確率密度関数は割愛
$$
N_n(\mu,\Sigma)
$$

## 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
$$
\left[ \begin{array}{ccc} 1.0 & 0.8\\ 0.8 & 1.0 \end{array} \right]
$$
多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。
### 《シード値について》
実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておきましょう。今回シード値は0を使うことにします。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# シード値を０に固定
np.random.seed(0)
# 平均
mu = [-3, 0]
# 共分散行列
sigma = [[1.0, 0.8], [0.8, 1.0]]
# 2次元正規乱数を500個作成
values = np.random.multivariate_normal(mu, sigma, 500)
```

## 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。<br>
散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
plt.scatter(values[:, 0],values[:, 1])
plt.title("scatter")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

## 【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。<br>
ヒストグラムはplt.hist()を使うことで描けます。<br>
各次元ごとに作成し、2つのヒストグラムとしてください。<br>
それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。<br>

```python
plt.hist(values[:, 0], bins=50, rwidth=1.0)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("X[0]")
plt.xlim(-6, 4)
plt.ylim(0, 35)
```

```python
plt.hist(values[:, 1], bins=50, rwidth=1.0)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("X[1]")
plt.xlim(-6,4)
plt.ylim(0, 35)
```

## 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
$$
\left[ \begin{array}{ccc} 1.0 & 0.8\\ 0.8 & 1.0 \end{array} \right]
$$
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題4のものは1を表示してください。

```python
# 平均
mu_y = [0, -3]
# 共分散行列
sigma_y = [[1.0, 0.8], [0.8, 1.0]]
# 2次元正規乱数を500個作成
values_y = np.random.multivariate_normal(mu_y, sigma_y, 500)
```

```python
plt.scatter(values[:, 0], values[:, 1], label="0")
plt.scatter(values_y[:, 0], values_y[:, 1], label="1")
plt.title("scatter")
plt.xlabel("x axis")
plt.ylabel("y axis")
# ラベルを表示させる
plt.legend()
plt.show()
```

## 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。<br>
結合はnp.concatenate()やnp.vstack()を使うことで行えます

```python
values_all = np.concatenate([values,values_y],0)
# 結合されているか散布図で確認
plt.scatter(values_all[:, 0], values_all[:, 1])
plt.title("scatter")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

## 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。<br>
問題1のものには0、問題4のものには1を対応させます。<br>
0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。<br>
機械学習に使用するデータセットはこのような形になっていることが多いです。<br>

```python
# y=xの一次関数で分類できるか確認
x = np.arange(-6, 3, 0.1)
y = x
plt.plot(x, y)
plt.scatter(values_all[:, 0], values_all[:, 1])
plt.title("scatter")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# ラベルを入れるための列を追加
values_all_label = np.insert(values_all, 2, 0, axis=1)
for val in values_all_label:
    if val[0]<val[1]:
        # 問題１の値ならラベル0
        val[2] = 0
    else:
        # 問題4の値ならラベル1
        val[2] = 1
```

```python
# 結果を確認する 問題1の値なら青、問題4の値ならオレンジで表示
for val in values_all_label:
    if val[2]==0:
        plt.scatter(val[0], val[1], c="b")
    else:
        plt.scatter(val[0], val[1], c="orange")
plt.title("scatter")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
